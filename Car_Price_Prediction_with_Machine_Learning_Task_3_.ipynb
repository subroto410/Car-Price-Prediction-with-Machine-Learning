```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df=pd.read_csv('/content/drive/MyDrive/dataset/car data.csv')
df
```

```python
input=df.drop(['Car_Name','Owner','Selling_type'],axis='columns')
target=df.Selling_Price
input
```

```python
from sklearn.preprocessing import LabelEncoder
Numerics=LabelEncoder()
```

```python
input['Fuel_Type']=Numerics.fit_transform(input['Fuel_Type'])
input['Transmission']=Numerics.fit_transform(input['Transmission'])

input
```

```python
input_s=input.drop(['Fuel_Type','Transmission','Selling_Price'],axis='columns')
input_s
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(input_s,target,test_size=0.2,random_state=0)
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
pred=model.predict(x_test)
print(pred)
```

```python
df
```

```python
import matplotlib.pyplot as plt
```

```python
training_data_prediction = model.predict(input_s)
```

```python
from sklearn import metrics
```

```python
error_score = metrics.r2_score(target,training_data_prediction)
print("R squared Error :",error_score)
```

```python
plt.scatter(target, training_data_prediction)
plt.xlabel("Actual price")
plt.ylabel("Predicted Price")
plt.title("Actual Prices vs Predicted Prices")
plt.show()
```
